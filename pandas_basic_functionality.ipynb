# Pandas - Basic Functionality

```python
# import pandas library
import pandas as pd
import numpy as np
```

### Series - Basic Functionality

```python
s = pd.Series(np.arange(1, 10, 2))
print(s)

print('\nThe axes are:', s.axes)
print('Is the object empty? -', s.empty)
print('Dimension:', s.ndim)
print('Size of the object:', s.size)
print('Values are:', s.values)
print('The first 2 rows:\n', s.head(2))
print('The last 2 rows:\n', s.tail(2))
```

### DataFrame - Basic Functionality

```python
#Create a Dictionary of series
data = {'Name':pd.Series(['Tom','James','Ricky','Vin','Steve','Smith','Jack']),
       'Age':pd.Series([25,26,25,23,30,29,23]),
       'Rating':pd.Series([4.23,3.24,3.98,2.56,3.20,4.6,3.8])}

df = pd.DataFrame(data)
print('Our data set is:')
print(df)

# Transpose of the dataset
print('\nTranspose of the data set:')
print(df.T)

# axes of dataset
print('\nRow and Column axes lebels are:')
print(df.axes)

# data types of each columns
print('\nData types of data set:')
print(df.dtypes)

# Checking, Is dataset empty or not?
print('\nIs data set empty? -', df.empty)

# dimension of dataset
print('\nDimension of dataset:')
print(df.ndim)

# shape of dataset
print('\nShape of dataset:')
print(df.shape)

# size of dataset
print('\nSize of dataset:')
print(df.size)

# values of dataset
print('\nValues of our DataFrame:')
print(df.values)

# show top 5 rows data
print('\nThe top 5 rows data:')
print(df.head())

# show bottom 5 rows data
print('\nThe bottom 3 rows data:')
print(df.tail(3))
```
