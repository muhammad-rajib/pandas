# Pandas - Sorting

```python
# import library
import pandas as pd
import numpy as np
```

```python
unsort_df = pd.DataFrame(np.random.randn(10, 2), index=[1,4,6,2,3,5,9,8,0,7], 
                        columns=['col1', 'col2'])
print(unsort_df, '\n')

# sort by label - ascending order
sorted_df = unsort_df.sort_index()
print('Sorted by row index:')
print(sorted_df, '\n')

# sort by label - descending order
sorted_df = unsort_df.sort_index(ascending=False)
print('Sorted by row index:')
print(sorted_df)
```

### Sort the Columns

```python
unsort_df = pd.DataFrame(np.random.randn(10, 2), index=[1,4,6,2,3,5,9,8,0,7],
                        columns=['col2', 'col1'])
print(unsort_df, '\n')

sorted_df = unsort_df.sort_index(axis=1)
print(sorted_df)
```

### Sort by Value

```python
unsorted_df = pd.DataFrame({'col1':[2,1,1,1],'col2':[1,3,2,4]})
print(unsorted_df, '\n')

sorted_df = unsorted_df.sort_values(by=['col1'])
print(sorted_df, '\n')

# sort by multiple columns
sorted_df = unsorted_df.sort_values(by=['col1', 'col2'])
print(sorted_df)
```
