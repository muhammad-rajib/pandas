## Exercise 3: Find the most expensive car company name.

```python
df1 = df[['company', 'price']]
most_expensive_car_company = df1.sort_values(by='price')
print(most_expensive_car_company[-1:])
```

## Exercise 4: Print All Toyota Cars details

## Exercise 6: Find each company’s Higesht price car

```python
car_manufacturers = df.groupby('company')
price_info = car_manufacturers[['company', 'price']].max()
price_info
```

## Exercise 7: Find the average mileage of each car making company

```python
car_manufacturers = df.groupby('company')
mileage_avg = car_manufacturers[['company', 'average-mileage']].mean()
mileage_avg
```

## Exercise 9: Concatenate two data frames using the following conditions

```python
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
JapaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}

cars_df1 = pd.DataFrame(GermanCars)
cars_df2 = pd.DataFrame(JapaneseCars)

cars_df = pd.concat([cars_df1, cars_df2], keys=['German', 'Japan'])
cars_df
```

## Exercise 10: Merge two data frames using the following condition

```python
CarPrice = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
CarHorsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}

cars_df1 = pd.DataFrame(CarPrice)
cars_df2 = pd.DataFrame(CarHorsepower)

cars_df = pd.merge(cars_df1, cars_df2, on='Company')
cars_df
```
