# Pandas - Data Structures

```python
# import pandas library
import pandas as pd
import numpy as np
```

## Pandas - Series
### pandas.Series(data, index, dtype, copy)

```python
# Create an Empty Series
s = pd.Series()
print(s)
```

```python
# Create a Series from ndarray
data = np.array(['a', 'b', 'c'])
s = pd.Series(data)
print(s)
```
