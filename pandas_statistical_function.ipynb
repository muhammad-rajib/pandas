## Pandas - Statistical Functions

### pct_change()

```python
# import required libraries
import pandas as pd
import numpy as np
```

```python
# percentage change
s = pd.Series([1,2,3,4,5,4])
print(s)

print(s.pct_change())
```

```python
# covariance of pair of list values
frame = pd.DataFrame(np.random.randn(10, 5), columns=['a', 'b', 'c', 'd', 'e'])
print(frame, '\n')

print(frame['a'].cov(frame['b']))
```

```python
# correlation
frame = pd.DataFrame(np.random.randn(10, 5), columns=['a', 'b', 'c', 'd', 'e'])
print(frame, '\n')

# correlation between column 'a' and 'b'
print(frame['a'].corr(frame['b']), '\n')

print(frame.corr())
```

```python
# Data Ranking
s = pd.Series(np.random.randn(5), index=list('abcde'))
print(s, '\n')

print(s.rank())
```
